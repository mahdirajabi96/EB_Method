```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Int_DF = pd.read_excel('SampleData.xlsx',sheet_name='Intersections')
Int_DF.index = Int_DF.Int_ID
Int_DF['TotCrashes'] = Int_DF.Tot_Crash_Y1+Int_DF.Tot_Crash_Y2+Int_DF.Tot_Crash_Y3
Crash_DF = pd.read_excel('SampleData.xlsx',sheet_name='Crashes')
Crash_DF.index = Crash_DF.Int_ID
Int_DF['Fatal'] = Crash_DF.Fatal.loc[Int_DF.index]
Int_DF['Injury'] = Crash_DF.Injury.loc[Int_DF.index]
Int_DF['PDO'] = Crash_DF.PDO.loc[Int_DF.index]
F1 = ['Int_ID', 'Traf_Ctrl', 'Leg_Count', 'AADT_Major', 'AADT_Minor', 'TotCrashes']
#df = pd.DataFrame(columns=list(set(Int_DF.AADT_Major)),index = list(set(Int_DF.AADT_Minor)))
display(Int_DF)
display(Crash_DF)
```

```python
#4.4.2.1 Crash Freq
Int_DF['CrashFreq'] = Int_DF['TotCrashes']/3.0
Int_DF.sort_values('CrashFreq',ascending=False)[F1 + ['CrashFreq']]
```

```python
#4.4.2.2 Crash Rate
Int_DF['MEV'] = Int_DF.AADT_Major + Int_DF.AADT_Minor
Int_DF['MEV'] = Int_DF['MEV']/1000000*3*365
Int_DF['CrashRate'] = Int_DF.TotCrashes/Int_DF.MEV
Int_DF.sort_values('CrashRate',ascending=False)[F1 + ['MEV','CrashRate']]
```

```python
#4.4.2.3 EPDO
Cost_DF = pd.DataFrame(columns=['Cost','Weight','Severity'])
Cost_DF.Severity = ['Fatal','Injury','PDO']
Cost_DF.index = Cost_DF.Severity
Cost_DF.Cost = [4008900,82600,7400]
Cost_DF.Weight = Cost_DF.Cost/Cost_DF.loc['PDO','Cost']
display(Cost_DF)
Int_DF['EPDO_Score'] = Int_DF.Fatal*Cost_DF.loc['Fatal','Weight']+Int_DF.Injury*Cost_DF.loc['Injury','Weight']+Int_DF.PDO*Cost_DF.loc['PDO','Weight']
Int_DF.sort_values('EPDO_Score',ascending=False)[F1 + ['EPDO_Score']]
```

```python
# 4.4.2.4 Relative Severity Index RSI
Cost_Type_DF = pd.DataFrame(columns=['Cost'],index = ['Rear_end_SG','Rear_end_ST','SS','Right_Angle_SG','Right_Angle_ST','Ped_Bike','Head_On_SG','Head_On_ST','FO','Other'])
Cost_Type_DF['Cost'] = [26700,13200,34000,47300,61100,158900,24100,47500,94700,55100]
Cost_Type_DF['Type'] = Cost_Type_DF.index.to_series()
display(Cost_Type_DF)
Int_DF['RSI_Sum'] = 0
for i in Int_DF[Int_DF.Traf_Ctrl=='Signal'].index:
    for c1,c2 in zip(['Rear_end_SG','SS','Right_Angle_SG','Ped_Bike','Ped_Bike','Head_On_SG','FO','Other'],
                     ['Rear_end'   ,'SS','Right_Angle'   ,'Ped'     ,'Bike'    ,'Head_On'   ,'FO','Other']):
        Int_DF.loc[i,'RSI_Sum'] += Crash_DF.loc[i,c2] * Cost_Type_DF.loc[c1,'Cost']
for i in Int_DF[Int_DF.Traf_Ctrl=='TWSC'].index:
    for c1,c2 in zip(['Rear_end_ST','SS','Right_Angle_ST','Ped_Bike','Ped_Bike','Head_On_ST','FO','Other'],
                     ['Rear_end'   ,'SS','Right_Angle'   ,'Ped'     ,'Bike'    ,'Head_On'   ,'FO','Other']):
        Int_DF.loc[i,'RSI_Sum'] += Crash_DF.loc[i,c2] * Cost_Type_DF.loc[c1,'Cost']
Int_DF['RSI_Ave'] = Int_DF['RSI_Sum'] / Int_DF.TotCrashes
#Int_DF['RSI_Ave'] = Int_DF['RSI_Ave'] // 100 * 100
Int_DF.loc[Int_DF.Traf_Ctrl=='Signal','RSI_Pop_Ave'] = Int_DF[Int_DF.Traf_Ctrl=='Signal']['RSI_Sum'].sum()/Int_DF[Int_DF.Traf_Ctrl=='Signal']['TotCrashes'].sum()
Int_DF.loc[Int_DF.Traf_Ctrl=='TWSC','RSI_Pop_Ave'] = Int_DF[Int_DF.Traf_Ctrl=='TWSC']['RSI_Sum'].sum()/Int_DF[Int_DF.Traf_Ctrl=='TWSC']['TotCrashes'].sum()
Int_DF['Exceed_Ave_RSI'] = Int_DF.RSI_Ave>Int_DF.RSI_Pop_Ave
Int_DF.sort_values('RSI_Ave',ascending=False)[F1 + ['RSI_Sum','RSI_Ave','RSI_Pop_Ave','Exceed_Ave_RSI']]
```

```python
# 4.4.2.5 Critical Rate
from scipy import stats
pvalue = stats.norm.ppf(.95)
Int_DF['WA_Crash_Rate'] =  pd.Series(Int_DF.groupby(Int_DF.Traf_Ctrl).apply(lambda DF:sum(DF.CrashRate*DF.MEV)/DF.MEV.sum()).loc[Int_DF.Traf_Ctrl].tolist(),index = Int_DF.index)
Int_DF['CriticalRate'] =  Int_DF.apply(lambda row:row.WA_Crash_Rate+pvalue*np.sqrt(row.WA_Crash_Rate/row.MEV)+(0.5/row.MEV),axis=1)
Int_DF['ExceedsCR'] = Int_DF.CriticalRate<Int_DF.CrashRate
Int_DF[F1 + ['MEV','CrashRate','WA_Crash_Rate','CriticalRate','ExceedsCR']]
```

```python
#4.4.2.6 Exess Predic. Ave. Crash Fq. using method of moments
Int_DF['G_Crash_Fq_A'] =  pd.Series(Int_DF.groupby(Int_DF.Traf_Ctrl).apply(lambda DF:DF.CrashFreq.mean()).loc[Int_DF.Traf_Ctrl].tolist(),index = Int_DF.index)
Int_DF['G_Crash_Fq_V'] =  pd.Series(Int_DF.groupby(Int_DF.Traf_Ctrl).apply(lambda DF:np.var(DF.CrashFreq,ddof=1)).loc[Int_DF.Traf_Ctrl].tolist(),index = Int_DF.index)
Int_DF['Crash_Fq_Adj'] = Int_DF.apply(lambda row:row.CrashFreq+row.G_Crash_Fq_A/row.G_Crash_Fq_V*(row.G_Crash_Fq_A-row.CrashFreq),axis=1)
Int_DF['PI'] = Int_DF.Crash_Fq_Adj - Int_DF.G_Crash_Fq_A
Int_DF.sort_values('PI',ascending=False)[F1 + ['CrashFreq','G_Crash_Fq_A','G_Crash_Fq_V','Crash_Fq_Adj','PI']]
```

```python
#4.4.2.7 LOSS
def SPF(row):
    if row.Traf_Ctrl == 'TWSC':
        a,b,c = -8.56,.6,.61
        cf = 1.7/4.06
    if row.Traf_Ctrl == 'Signal':
        a,b,c = -5.13,.6,.2
        cf = 0.4
    a_j = np.log(row.AADT_Major)
    a_n = np.log(row.AADT_Minor)
    spf = np.exp(a+b*a_j+c*a_n)
    return(spf*cf)
Int_DF['N_SPF'] = Int_DF.apply(SPF,axis=1)
Int_DF['k'] = pd.Series(pd.Series([0.4,0.11],index=['TWSC','Signal']).loc[Int_DF.Traf_Ctrl].tolist(),index=Int_DF.index)
Int_DF['SPF_STD'] = Int_DF.apply(lambda row:np.sqrt(row.k*row.N_SPF**2),axis=1)
Int_DF['LOSS I'] = Int_DF.apply(lambda row:pd.Interval(0,row.N_SPF-1.5*row.SPF_STD,'left'),axis=1)
Int_DF['LOSS II'] = Int_DF.apply(lambda row:pd.Interval(row.N_SPF-1.5*row.SPF_STD,row.N_SPF,'left'),axis=1)
Int_DF['LOSS III'] = Int_DF.apply(lambda row:pd.Interval(row.N_SPF,row.N_SPF+1.5*row.SPF_STD,'left'),axis=1)
Int_DF['LOSS IV'] = Int_DF.apply(lambda row:pd.Interval(row.N_SPF+1.5*row.SPF_STD,np.Inf,'left'),axis=1)
Int_DF['LOSS'] = Int_DF.apply(lambda row:list(set([loss if row.CrashFreq in row[loss] else 0 for loss in ['LOSS I','LOSS II', 'LOSS III', 'LOSS IV']]))[-1],axis=1)
Int_DF.sort_values('LOSS',ascending=False)[F1 + ['CrashFreq','N_SPF','SPF_STD','LOSS']]
```

```python
# 4.4.2.8 Excess Predicted Average Crash Fq.
Int_DF['ExcessPredCF'] = Int_DF.CrashFreq - Int_DF.N_SPF
Int_DF.sort_values('ExcessPredCF',ascending=False)[F1 + ['CrashFreq','N_SPF','ExcessPredCF']]
```
